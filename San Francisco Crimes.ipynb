# 1 Crimes por categoria

```python
crime_cat = dados[['IncidntNum', 'Category']].groupby('Category').count().reset_index().sort_values('IncidntNum', ascending=False)
crime_cat.columns = ['Category', 'Count']
crime_cat = crime_cat.reset_index(drop=True)
crime_cat

```

# 2 Total de crimes por bairro 

```python
crime_dis = dados[['IncidntNum', 'PdDistrict']].groupby('PdDistrict').count().sort_values('IncidntNum', ascending=False)
crime_dis = crime_dis.reset_index(drop=False)
crime_dis
```

# 3 Crimes por dia da semana

```python
crime_day = dados[['IncidntNum', 'DayOfWeek', 'Category']].groupby(['DayOfWeek']).count()
crime_day = crime_day.reset_index(drop=False)
crime_day
```

# 3.1 Crimes mais ocorridos por dia da semana
